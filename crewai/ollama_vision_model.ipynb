```python
from IPython.display import Image
#Let's visualize the image we are going to ask VLmodels to read
img_file = 'NTP_screenshot_1.png'
Image(filename=img_file)
```

```python
import ollama

'''
small models to try
gemma3:4b (gemma3 has smaller models; however, 4b is the smallest vision model.)
qwen2.5vl:3b (smallest vision model among qwen2.5vl series. Recently, qwen3-VL is released, which is much bigger model.)
'''

result = ollama.chat(model='gemma3:4b',
                         messages=[
                            {
                                'role':'user',
                                'content':'Describe this image:',
                                'images':[img_file]
                            }
                         ]
                    )
```

```python
print(result)
```

```python
print(result['message']['content'])
```
